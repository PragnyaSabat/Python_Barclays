# Task
The task is to identify dormant accounts from the transactional data (`df1`) by calculating the maximum interval between consecutive transactions for each `AccountID`. Accounts with a maximum transaction interval of two months or more will be flagged as 'Dormant', otherwise 'Active'. Finally, display the dormant accounts along with their inactivity periods and the updated DataFrame with the dormancy flags.

## Calculate Transaction Intervals

### Subtask:
For each AccountID, calculate the time difference between consecutive TransactionDate entries. This will require sorting by TransactionDate within each account group.


**Reasoning**:
To calculate the time difference between consecutive transactions for each account, I need to first sort the DataFrame by `AccountID` and then by `TransactionDate`. After sorting, I will group by `AccountID` and apply the `.diff()` method to the `TransactionDate` column to find the time intervals.



```python
df1 = df1.sort_values(by=['AccountID', 'TransactionDate']).reset_index(drop=True)
df1['TransactionInterval'] = df1.groupby('AccountID')['TransactionDate'].diff()
print(df1[['AccountID', 'TransactionDate', 'TransactionInterval']].head(20))
```

**Reasoning**:
Now that the transaction intervals are calculated, I need to find the maximum `TransactionInterval` for each `AccountID` to determine the longest period of inactivity. This will involve grouping by `AccountID` and then taking the maximum of the `TransactionInterval` column.



```python
max_inactivity = df1.groupby('AccountID')['TransactionInterval'].max().dt.days.reset_index()
max_inactivity.rename(columns={'TransactionInterval': 'MaxInactivityDays'}, inplace=True)
print(max_inactivity.head())
```

**Reasoning**:
Now that the maximum inactivity period for each account has been calculated, I will merge this information back into the original DataFrame (`df1`) and then classify accounts as 'Dormant' or 'Active' based on whether their maximum inactivity period is two months (60 days) or more. Finally, I will display the dormant accounts along with their inactivity periods and show the updated DataFrame with the new dormancy flags.



```python
df1 = df1.merge(max_inactivity, on='AccountID', how='left')
df1['DormancyStatus'] = df1['MaxInactivityDays'].apply(lambda x: 'Dormant' if x >= 60 else 'Active')
dormant_accounts = df1[df1['DormancyStatus'] == 'Dormant'].drop_duplicates(subset=['AccountID'])[['AccountID', 'MaxInactivityDays', 'DormancyStatus']]
print("Dormant Accounts and their Inactivity Periods:")
display(dormant_accounts.head())
print("\nDataFrame with Dormancy Flags:")
print(df1[['AccountID', 'TransactionDate', 'TransactionInterval', 'MaxInactivityDays', 'DormancyStatus']].head(20))
```

# Task
Calculate the total number of transactions for each account, define rubrics for 'High', 'Medium', and 'Low' activity levels based on these frequencies, assign an activity level to each account, and then summarize the distribution of accounts across these levels, including the defined rubrics.

## Calculate Transaction Frequency

### Subtask:
Calculate the total number of transactions for each account to determine their frequency of activity.


**Reasoning**:
To calculate the total number of transactions for each account, I need to group the DataFrame `df1` by 'AccountID' and then count the occurrences of each account.



```python
transaction_frequency = df1.groupby('AccountID').size().reset_index(name='TransactionCount')
print(transaction_frequency.head())
```

**Reasoning**:
Now that the transaction frequencies are calculated, I need to define rubrics for 'High', 'Medium', and 'Low' activity levels. A common approach is to use quantiles to define these thresholds, ensuring a balanced distribution of accounts across the activity levels. Then, I will assign an activity level to each account based on these rubrics.



```python
quantiles = transaction_frequency['TransactionCount'].quantile([0.33, 0.66])
low_threshold = quantiles[0.33]
high_threshold = quantiles[0.66]

def assign_activity_level(count):
    if count <= low_threshold:
        return 'Low'
    elif count <= high_threshold:
        return 'Medium'
    else:
        return 'High'

transaction_frequency['ActivityLevel'] = transaction_frequency['TransactionCount'].apply(assign_activity_level)
print("Activity Level Rubrics (Transaction Count):")
print(f"Low: <= {low_threshold:.0f}")
print(f"Medium: > {low_threshold:.0f} and <= {high_threshold:.0f}")
print(f"High: > {high_threshold:.0f}")
print("\nTransaction Frequency with Activity Levels (head):")
print(transaction_frequency.head())
```

**Reasoning**:
To summarize the distribution of accounts across the newly assigned activity levels, I will calculate the count of accounts for each 'ActivityLevel' and display the results.



```python
activity_level_distribution = transaction_frequency['ActivityLevel'].value_counts()
print("\nDistribution of Accounts by Activity Level:")
print(activity_level_distribution)
```

Establish rubrics (thresholds) for 'High', 'Medium', and 'Low' for average balance, likely using quantiles to ensure a balanced distribution.

## Identify Large Withdrawals

### Subtask:
Define a threshold for 'large' withdrawals and identify transactions in the 'df1' DataFrame where 'TransactionType' is 'Withdrawal' and 'TransactionAmount' exceeds this threshold. Create a new column to flag these transactions.


**Reasoning**:
To define a threshold for 'large' withdrawals, I will first filter the DataFrame to include only withdrawal transactions and then examine the distribution of their absolute transaction amounts using descriptive statistics.



```python
withdrawal_amounts = df1[df1['TransactionType'] == 'Withdrawal']['TransactionAmount'].abs()
print("Descriptive statistics for withdrawal amounts:")
print(withdrawal_amounts.describe())

# Based on the descriptive statistics, let's choose a threshold for 'large' withdrawals.
# For example, we can consider amounts above the 75th percentile or a fixed value.
# Let's use the 75th percentile as the threshold for 'large' withdrawals.
large_withdrawal_threshold = withdrawal_amounts.quantile(0.75)
print(f"\nChosen threshold for large withdrawals (75th percentile): {large_withdrawal_threshold:.2f}")

df1['IsLargeWithdrawal'] = ((df1['TransactionType'] == 'Withdrawal') &
                            (df1['TransactionAmount'].abs() > large_withdrawal_threshold))

print("\nFirst 5 transactions flagged as large withdrawals:")
print(df1[df1['IsLargeWithdrawal']].head())
```

**Reasoning**:
Now that large withdrawals and overdrafts are flagged, the next step is to calculate the frequency of these events for each unique `AccountID` to understand which accounts are more prone to them.



```python
large_withdrawal_frequency = df1.groupby('AccountID')['IsLargeWithdrawal'].sum().reset_index(name='LargeWithdrawalCount')
overdraft_frequency = df1.groupby('AccountID')['IsOverdraft'].sum().reset_index(name='OverdraftCount')

print("Frequency of Large Withdrawals per Account (head):")
print(large_withdrawal_frequency.head())
print("\nFrequency of Overdrafts per Account (head):")
print(overdraft_frequency.head())
```

# Task
Calculate balance volatility using standard deviation or coefficient of variation.

**Reasoning**:
To calculate balance volatility, I will group the `df1` DataFrame by `AccountID`. For each account, I will compute the standard deviation of `AccountBalance` to measure absolute volatility and the mean of `AccountBalance` to calculate the coefficient of variation (CV), which normalizes volatility by the average balance, making it comparable across accounts.

```python
balance_volatility = df1.groupby('AccountID')['AccountBalance'].agg(['std', 'mean']).reset_index()
balance_volatility.rename(columns={'std': 'BalanceStdDev', 'mean': 'AverageAccountBalance'}, inplace=True)

# Calculate Coefficient of Variation (CV)
# CV = (Standard Deviation / Mean) * 100. Handle cases where mean is zero to avoid division by zero.
balance_volatility['BalanceCV'] = balance_volatility.apply(
    lambda row: (row['BalanceStdDev'] / row['AverageAccountBalance']) * 100
                if row['AverageAccountBalance'] != 0 else 0, axis=1
)

print("Balance Volatility per Account (head):")
print(balance_volatility.head())
```

## Calculate IQR for BalanceCV

### Subtask:
Compute the first quartile (Q1), third quartile (Q3), and the Interquartile Range (IQR) for the 'BalanceCV' column in the 'balance_volatility' DataFrame.


**Reasoning**:
To calculate the first quartile (Q1), third quartile (Q3), and the Interquartile Range (IQR) for the 'BalanceCV' column, I will use the .quantile() method for Q1 and Q3, and then calculate the difference to get the IQR, and finally print the results.



```python
Q1_cv = balance_volatility['BalanceCV'].quantile(0.25)
Q3_cv = balance_volatility['BalanceCV'].quantile(0.75)
IQR_cv = Q3_cv - Q1_cv

print(f"First Quartile (Q1) for BalanceCV: {Q1_cv:.2f}")
print(f"Third Quartile (Q3) for BalanceCV: {Q3_cv:.2f}")
print(f"Interquartile Range (IQR) for BalanceCV: {IQR_cv:.2f}")
```

**Reasoning**:
Now that the quartiles and IQR for 'BalanceCV' have been calculated, I need to define the upper and lower bounds for identifying outliers using the IQR method. This will involve multiplying the IQR by a constant (typically 1.5) and adding/subtracting it from Q3 and Q1 respectively.



```python
lower_bound_cv = Q1_cv - 1.5 * IQR_cv
upper_bound_cv = Q3_cv + 1.5 * IQR_cv

print(f"\nLower Bound for BalanceCV (IQR method): {lower_bound_cv:.2f}")
print(f"Upper Bound for BalanceCV (IQR method): {upper_bound_cv:.2f}")
```

**Reasoning**:
Now that the IQR bounds are defined, I will identify the anomalous accounts by filtering the `balance_volatility` DataFrame for 'BalanceCV' values that fall outside these bounds.



```python
iqr_anomalies = balance_volatility[
    (balance_volatility['BalanceCV'] < lower_bound_cv) |
    (balance_volatility['BalanceCV'] > upper_bound_cv)
]
print("Anomalous Accounts (IQR method) based on BalanceCV:")
print(iqr_anomalies)
```

**Reasoning**:
To identify anomalous accounts using the Z-score method, I first need to calculate the mean and standard deviation of the 'BalanceCV' column. Then, I will compute the Z-score for each 'BalanceCV' value.



```python
mean_cv = balance_volatility['BalanceCV'].mean()
std_cv = balance_volatility['BalanceCV'].std()

balance_volatility['ZScore_CV'] = (balance_volatility['BalanceCV'] - mean_cv) / std_cv

print(f"Mean of BalanceCV: {mean_cv:.2f}")
print(f"Standard Deviation of BalanceCV: {std_cv:.2f}")
print("\nBalance Volatility with Z-Scores (head):")
print(balance_volatility.head())
```

```python
zscore_threshold = 2.0 # Common threshold for anomalies
zscore_anomalies = balance_volatility[
    (balance_volatility['ZScore_CV'].abs() > zscore_threshold)
]
print(f"Anomalous Accounts (Z-score method) based on BalanceCV (threshold = {zscore_threshold}):")
print(zscore_anomalies)
```

### Subtask:
Conduct an independent samples t-test on the 'AverageAccountBalance' between the 'high-volume' and 'low-volume' account groups to statistically compare their means. This will help determine if there is a significant difference in average balances between the two groups. Include the p-value and t-statistic in the output.


**Reasoning**:
To statistically compare the average account balances of high-volume and low-volume accounts, I will conduct an independent samples t-test. This requires importing the `ttest_ind` function from `scipy.stats`, extracting the relevant data, and then executing the test to obtain the t-statistic and p-value.



```python
from scipy.stats import ttest_ind

# Extract 'AverageAccountBalance' for high-volume accounts
high_volume_balance = high_volume_accounts['AverageAccountBalance']

# Extract 'AverageAccountBalance' for low-volume accounts
low_volume_balance = low_volume_accounts['AverageAccountBalance']

# Perform independent samples t-test
t_statistic, p_value = ttest_ind(high_volume_balance, low_volume_balance, equal_var=False) # Assuming unequal variances

print(f"Independent Samples t-test Results:")
print(f"  T-statistic: {t_statistic:.3f}")
print(f"  P-value: {p_value:.3f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Prepare data for plotting
plot_data = pd.DataFrame({
    'ActivityLevel': ['High-Volume'] * len(high_volume_balance) + ['Low-Volume'] * len(low_volume_balance),
    'AverageAccountBalance': pd.concat([high_volume_balance, low_volume_balance])
})

plt.figure(figsize=(8, 6))
sns.barplot(x='ActivityLevel', y='AverageAccountBalance', data=plot_data, palette={'High-Volume': 'skyblue', 'Low-Volume': 'lightcoral'})
plt.title('Average Account Balance: High-Volume vs. Low-Volume Accounts')
plt.xlabel('Activity Level')
plt.ylabel('Average Account Balance')

# Annotate the plot with t-statistic and p-value
plt.text(0.5, plot_data['AverageAccountBalance'].max() * 0.95, f'T-statistic: {t_statistic:.3f}',
         horizontalalignment='center', color='black', fontsize=12)
plt.text(0.5, plot_data['AverageAccountBalance'].max() * 0.90, f'P-value: {p_value:.3f}',
         horizontalalignment='center', color='black', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
high_volume_accounts = account_metrics[account_metrics['ActivityLevel'] == 'High']
low_volume_accounts = account_metrics[account_metrics['ActivityLevel'] == 'Low']

print("High-Volume Accounts (head):")
display(high_volume_accounts.head())
print(f"Shape of High-Volume Accounts: {high_volume_accounts.shape}")

print("\nLow-Volume Accounts (head):")
display(low_volume_accounts.head())
print(f"Shape of Low-Volume Accounts: {low_volume_accounts.shape}")
```

**Reasoning**:
To statistically compare the average account balances of high-volume and low-volume accounts using a Z-test, I will import the `ztest` function from `statsmodels.stats.weightstats`, extract the 'AverageAccountBalance' from both filtered DataFrames, and then execute the Z-test to obtain the Z-statistic and p-value.



```python
from statsmodels.stats.weightstats import ztest

# Extract 'AverageAccountBalance' for high-volume accounts
high_volume_balance = high_volume_accounts['AverageAccountBalance']

# Extract 'AverageAccountBalance' for low-volume accounts
low_volume_balance = low_volume_accounts['AverageAccountBalance']

# Perform independent samples Z-test
z_statistic, p_value_z = ztest(high_volume_balance, low_volume_balance, alternative='two-sided')

print(f"Independent Samples Z-test Results:")
print(f"  Z-statistic: {z_statistic:.3f}")
print(f"  P-value: {p_value_z:.3f}")
```

## Interpret Z-test Results

The Z-test was conducted to compare the 'AverageAccountBalance' between 'high-volume' accounts and 'low-volume' accounts.

### Z-test Results:
*   **Z-statistic**: -0.942
*   **P-value**: 0.346

### Interpretation:

The p-value obtained from the Z-test is 0.346. This value is greater than the commonly used significance level of 0.05.

### Conclusion:

Therefore, we **do not reject the null hypothesis**. This indicates that there is **no statistically significant difference** in the 'AverageAccountBalance' between 'high-volume' accounts and 'low-volume' accounts in this dataset.

```python
df1['TransactionDate'] = pd.to_datetime(df1['TransactionDate'], errors='coerce')
monthly_segment_trends = df1.groupby([df1['TransactionDate'].dt.to_period('M'), 'CustomerSegment'])[['Credits', 'Debits']].sum().reset_index()
monthly_segment_trends['TransactionDate'] = monthly_segment_trends['TransactionDate'].dt.to_timestamp()

print("Monthly Credits and Debits by Customer Segment (head):")
print(monthly_segment_trends.head())
```

```python
unique_segments = monthly_segment_trends['CustomerSegment'].unique()

for segment in unique_segments:
    segment_data = monthly_segment_trends[monthly_segment_trends['CustomerSegment'] == segment]
    plt.figure(figsize=(14, 7))
    sns.lineplot(x='TransactionDate', y='Credits', data=segment_data, label='Total Credits')
    sns.lineplot(x='TransactionDate', y='Debits', data=segment_data, label='Total Debits')
    plt.title(f'Monthly Trends in Total Credits vs. Debits for {segment}')
    plt.xlabel('Date')
    plt.ylabel('Amount')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
